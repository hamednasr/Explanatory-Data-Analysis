## Aggregating and Ranking

Create the following __Seaborn Heatmap__ with __Medal Ranks__ for Top 50 Countries (Total Medals, Summer Games Medals, Winter Games Medals, Men, Women).

```python
medal_rank1= olympics.pivot_table(index='Country',
                     columns='Gender',
                     values='Athlete',
                     aggfunc='count',
                     margins=True,
                     margins_name='Total',
                     fill_value=0).\
                     sort_values('Total',ascending=False).drop('Total',axis=0)
medal_rank1 = medal_rank1[['Total','Women','Men']]
medal_rank1
```

```python
medal_rank2= olympics.pivot_table(index='Country',
                     columns='Edition',
                     values='Athlete',
                     aggfunc='count',
                     margins=True,
                     margins_name='Total',
                     fill_value=0).\
                     sort_values('Total',ascending=False).\
                     drop('Total',axis=0).drop('Total',axis=1)
medal_rank2 = medal_rank2[['Winter','Summer']]
medal_rank2
```

```python
ranked_medals =  pd.concat([medal_rank1,medal_rank2],axis=1).rank(ascending=False)
```

```python
ranked_medals.rank()
```

```python
plt.figure(figsize=(20,7))
sns.heatmap(data= ranked_medals.head(30).T,annot=True,cbar=False,cmap='RdYlGn_r')
```
